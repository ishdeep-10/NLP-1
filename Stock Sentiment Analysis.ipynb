```python
train = data[data['Date'] < '20150101']
test = data[data['Date'] > '20141231']
```

### Cleaning Data and Feature Engineering

```python
clean_data = train.iloc[:,2:27]
clean_data.replace('[^a-zA-Z]',' ',regex=True,inplace=True)

#Removing column names and replcing with indexes

list1 = [i for i in range(25)]
new_index = [str(i) for i in list1]
clean_data.columns = new_index
```

```python
clean_data.head()
```

```python
for index in new_index:
    clean_data[index] = clean_data[index].str.lower()
```

```python
clean_data.head(1)
```

```python
### Joining all the news from a date into a single headline

headlines=[]
for row in range(0,len(clean_data.index)):
    headlines.append(' '.join(str(x) for x in clean_data.iloc[row,0:25]))
```

```python
headlines[1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range=(2,2))
traindata = countvector.fit_transform(headlines)
```

```python
randomclassifier = RandomForestClassifier(n_estimators=200 , criterion='entropy')
randomclassifier.fit(traindata,train['Label'])
```

```python
### For Test Data

test_headlines=[]
for row in range(0,len(test.index)):
    test_headlines.append(' '.join(str(x) for x in test.iloc[row,2:27]))
```

```python
testdata = countvector.transform(test_headlines)
pred = randomclassifier.predict(testdata)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix = confusion_matrix(test['Label'],pred)
print(matrix)
```

```python
score = accuracy_score(test['Label'],pred)
print(score)
```

```python
report = classification_report(test['Label'],pred)
print
```
